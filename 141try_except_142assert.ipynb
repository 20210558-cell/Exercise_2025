```python
intList = range(10)
try:
  idx = int(input('숫자를 입력하세요.'))
  print(intList[idx])
except ValueError:
  print('입력한 내용이 아닙니다!')
except IndexError as e:
  print(f'인덱스 오류가 발생: {e}')
```

```python
data = ['a', 'bc', 1234, 'def', 'ghij']
try:
  for d in data:
    assert type(d) is str, f'문자열이 아닌 데이터 <{d}> 있음!'
    print(d)
except AssertionError as e:
  print(e)
```
