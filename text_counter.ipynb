```python
def count_characters(text):
    total_with_space = len(text)
    total_no_space = 0

    kor_count = 0
    eng_count = 0
    num_count = 0
    etc_count = 0

    for char in text:
        if not char.isspace():
            total_no_space += 1
        if '가' <= char <= '힣':
            kor_count += 1
        elif char.isascii() and char.isalpha():
            eng_count += 1
        elif char.isdigit():
            num_count += 1
        elif not char.isspace():
            etc_count += 1

    print("\n" + "="*30)
    print(f"공백 포함   : {total_with_space}자")
    print(f"공백 미포함 : {total_no_space}자")
    print("-" * 30)
    print(f"한글 : {kor_count}자")
    print(f"영어 : {eng_count}자")
    print(f"숫자 : {num_count}자")
    print(f"기타(특수문자 등) : {etc_count}자")
    print("="*30 + "\n")

print("=== 글자 수 카운터 ===")
print("(결과를 출력하려면 엔터를 두 번 입력하세요.)")

while True:
    print("\n텍스트를 입력해주세요. (종료: 's' 입력)")

    lines = []
    stop_signal = False

    while True:
        try:
            line = input()
            if line.strip().lower() == 's' and not lines:
                print("\n프로그램 종료")
                stop_signal = True
                break

            if not line:
                break
            lines.append(line)
        except EOFError:
            break

    if stop_signal:
        break

    if not lines:
        continue

    full_text = "\n".join(lines)
    count_characters(full_text)
```
