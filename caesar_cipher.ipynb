```python
print("=== 문자 암호화 프로그램 ===")

text = input("메시지 입력: ")
encrypted_text = ""

for char in text:
    if char == " ":
        encrypted_text += " "
    else:
        code_num = ord(char)
        new_code = code_num + 99
        new_char = chr(new_code)
        encrypted_text += new_char

print("\n[암호화 결과]")
print(encrypted_text)
```

```python
print("=== 문자 해석 프로그램 ===")

text = input("메시지 입력: ")
decryption_text = ""

for char in text:
    if char == " ":
        decryption_text += " "
    else:
        code_num = ord(char)
        new_code = code_num - 99
        new_char = chr(new_code)
        decryption_text += new_char

print("\n[암호 해석 결과]")
print(decryption_text)
```
