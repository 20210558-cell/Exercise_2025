```python
print("=== 환율 계산기 ===")
print("1. 원화(KRW) -> 달러(USD)")
print("2. 달러(USD) -> 원화(KRW)")

exchange_rate = 1467

try:
    choice = input("변환할 메뉴를 선택하세요 (1 또는 2): ")

    if choice == "1":
        won = int(input("환전할 금액(원)을 입력하세요: "))
        usd = won / exchange_rate
        print("------------------------------")
        print(f"결과: ${usd:.2f} (약 {int(usd)}달러)")

    elif choice == "2":
        usd = float(input("환전할 금액(달러)을 입력하세요: "))
        won = usd * exchange_rate
        print("------------------------------")
        print(f"결과: {format(int(won), ',')}원")

    else:
        print("잘못된 입력입니다. 1번이나 2번을 눌러주세요.")

except ValueError:
    print("오류: 금액은 숫자로만 입력해주세요.")
```
