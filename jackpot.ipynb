```python
import random
import time

money = 1000
bet_money = 100

print("=== 숫자 잭팟 게임 (숫자 3개가 똑같으면 잭팟) ===")
print(f"시작 소지금: {money}원")

while money >= bet_money:
    user_input = input(f"\n[잔액: {money}원] 엔터를 누르면 시작! (종료: s) > ")

    if user_input == "s":
        break

    money -= bet_money

    print("숫자 뽑는중..", end="")
    time.sleep(0.5)

    n1 = random.randint(1, 9)
    n2 = random.randint(1, 9)
    n3 = random.randint(1, 9)

    print(f"\n   [ {n1} ] [ {n2} ] [ {n3} ]   ")

    if n1 == n2 == n3:
        hit = 10000
        print(f"잭팟!! {hit}원 당첨!")
        money += hit

    elif n1 == n2 or n2 == n3 or n1 == n3:
        hit = 200
        print(f"숫자 2개 적중: {hit}원 당첨")
        money += hit

    else:
        print("꽝.. 다음 기회에")

print("----------------------------")
print(f"게임 종료. 최종 잔액: {money}원")
```
