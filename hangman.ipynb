```python
import random

hangman_pics = [
    """
      +---+
      |   |
          |
          |
          |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========
    """,
    """
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    =========
    """
]

word_list = ["one", "two", "three", "four", "five"]

print("=== 행맨 게임 ===")\

chosen_word = random.choice(word_list).upper()
word_length = len(chosen_word)
display = ["_"] * word_length
lives = 6
guessed_letters = []

game_over = False

while not game_over:
    print(hangman_pics[6 - lives])
    print(f"남은 목숨: {lives}")
    print(f"현재 단어: {' '.join(display)}")
    print("-" * 30)

    guess = input("알파벳 한 글자를 입력하세요 : ").upper()

    if len(guess) != 1 or not guess.isalpha():
        print("알파벳 한 글자만 입력해주세요.")
        continue

    if guess in guessed_letters:
        print(f"이미 입력한 글자입니다: {guess}")
        continue

    guessed_letters.append(guess)

    if guess in chosen_word:
        print(f"정답! '{guess}'가 포함되어 있습니다.")

        for i in range(word_length):
            if chosen_word[i] == guess:
                display[i] = guess
    else:
        print(f"틀렸습니다. '{guess}'는 없습니다.")
        lives -= 1

    if "_" not in display:
        game_over = True
        print("\n" + "="*30)
        print(f"정답! '{chosen_word}'였습니다.")
        print("="*30)

    if lives == 0:
        game_over = True
        print(hangman_pics[6])
        print("\n" + "="*30)
        print("게임 오버..")
        print(f"정답은 '{chosen_word}'였습니다.")
        print("="*30)
```
