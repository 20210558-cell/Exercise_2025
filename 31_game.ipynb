```python
import random
import time

print("=== 배스킨라빈스 31 게임  ===")
print("31을 외치는 사람이 패배합니다.")
print("한 번에 1개에서 3개의 숫자를 부를 수 있습니다.")

current_num = 0

while True:
    print(f"\n[내 차례] 현재 숫자: {current_num}")

    while True:
        try:
            my_count = int(input("숫자 몇 개를 부를까요? (1~3) : "))
            if 1 <= my_count <= 3:
                break #
            else:
                print("1, 2, 3 중에서만 입력하세요.")
        except ValueError:
            print("숫자로 입력해주세요.")

    for i in range(my_count):
        current_num += 1
        print(f"나: {current_num}")

        if current_num >= 31:
            break

    if current_num >= 31:
        print("\n패배했습니다..")
        break

    com_count = random.randint(1, 3)

    for i in range(com_count):
        current_num += 1
        print(f"컴퓨터: {current_num}")

        if current_num >= 31:
            break

    if current_num >= 31:
        print("\n승리!")
        break
```
