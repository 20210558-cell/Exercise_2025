```python
data1 = 'This is data1'
data2 = 'This is data2'
with open('mydata.txt', 'w') as f:
  f.writelines(data1)
  f.writelines(data2)
```

```python
with open('mydata.txt', 'r') as f:
  data = f.read()
print(data)
```

```python
data = ['This is data1\n', 'This is data2\n']
with open('mydata.txt', 'w') as f:
  f.writelines(data)
```

```python
with open('mydata.txt', 'r') as f:
  data = f.read()
print(data)
```

```python
bufsize = 1024                                 # 1K 바이트
with open('sunrise.jpg', 'rb') as f:
  with open('sunrise_copied.jpg', 'wb') as h:
    data = f.read(bufsize)                     # 'sunrise.jpg'를 1KB씩 읽어 들임
    while data:
      h.write(data)                            # 읽은 1KB 데이터를 씀
      data = f.read(bufsize)                   # 'sunrise.jpg'를 1KB씩 읽어 들임
```
