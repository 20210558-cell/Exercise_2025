```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
others = {'제갈량':99, '황충':70}
greats.update(others)
print(greats)
```

```python
greats.update(유비=90, 관우=96)
print(greats)
```

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
others = {'제갈량':99, '황충':70}
updated = greats | others
print(updated)
```

```python
updated2 = updated | {'유비':90, '관우':96}
print(updated2)
```

```python
greats |= others    # greats.update(others)와 동일
print(greats)
```
