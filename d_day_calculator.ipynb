```python
from datetime import datetime

def d_day_calculator():
    print("=== D-Day 계산기 ===")

    while True:
        target_input = input("\n날짜 입력 (예: 2025-12-17) 또는 's' 종료 > ")

        if target_input.lower() == 's':
            print("\n프로그램 종료")
            break

        try:
            target_date = datetime.strptime(target_input, "%Y-%m-%d")
            today = datetime.now()

            diff = target_date - today
            days = diff.days + 1

            print("-" * 30)
            if days > 0:
                print(f"[D-{days}] 그 날까지 {days}일 남았습니다.")
            elif days == 0:
                print(f"[D-Day] 오늘입니다.")
            else:
                print(f"[D+{abs(days)}] 그 날로부터 {abs(days)}일 지났습니다.")
            print("-" * 30)

        except ValueError:
            print("날짜 형식이 올바르지 않습니다. (예: 2025-12-17)")

d_day_calculator()
```
