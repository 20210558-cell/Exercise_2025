```python
import requests

def get_clean_weather(city_name):
    custom_format = "%l|%C|%t|%h|%w"

    url = f"https://wttr.in/{city_name}?format={custom_format}&lang=ko&M"

    try:
        response = requests.get(url)

        if response.status_code == 200:
            text_data = response.text.strip()

            if "Unknown location" in text_data:
                print(f"⚠️ '{city_name}'이라는 지역을 찾을 수 없습니다.")
                return

            parts = text_data.split("|")

            if len(parts) == 5:
                location, condition, temp, humidity, wind = parts

                print(f"\n======== {location} 날씨 정보 ========")
                print(f"날씨: {condition}")
                print(f"기온: {temp}")
                print(f"습도: {humidity}")
                print(f"바람: {wind}")
                print("=======================================")
            else:
                print(f"정보: {text_data}")
        else:
            print("날씨 정보를 가져오는 데 실패했습니다.")

    except Exception as e:
        print(f"오류 발생: {e}")

print("=== 날씨 정보 ===")

while True:
    city = input("\n지역을 입력해주세요. (종료: s) : ")

    if city.lower() == 's':
        print("\n프로그램 종료")
        break

    if not city:
        continue

    get_clean_weather(city)
```
