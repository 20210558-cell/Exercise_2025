```python
import random

menu = ["가위", "바위", "보"]
my_score = 0
com_score = 0

print("=== 가위바위보 게임 (삼세판) ===")
while my_score < 2 and com_score < 2:

    print(f"\n[현재 스코어] 나: {my_score} vs 컴퓨터: {com_score}")

    computer_choice = random.choice(menu)
    user_choice = input("가위, 바위, 보 중 하나를 입력하세요: ")

    if user_choice not in menu:
        print("잘못된 입력입니다.")
        continue

    print(f"나: {user_choice}")
    print(f"컴퓨터: {computer_choice}")

    if user_choice == computer_choice:
        print("결과: 비겼습니다! (점수 변동 없음)")

    elif (user_choice == "가위" and computer_choice == "보") or \
         (user_choice == "바위" and computer_choice == "가위") or \
         (user_choice == "보" and computer_choice == "바위"):
        print("결과: 이겼습니다!")
        my_score += 1

    else:
        print("결과: 졌습니다.")

print("\n============================")
if my_score == 2:
    print(f"최종 승리!({my_score} 대 {com_score})")
else:
    print(f"최종 패배({my_score} 대 {com_score})")
```
