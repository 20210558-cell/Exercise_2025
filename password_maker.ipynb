```python
import random

letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "0123456789"
specials = "!@#$%"

all_chars = letters + numbers + specials

print("=== 랜덤 비밀번호 생성기 ===")
length = int(input("몇 글자의 비밀번호를 만들까요? (예: 12) : "))

password = ""

for i in range(length):
    password += random.choice(all_chars)

print("\n---------------------------")
print(f"생성된 비밀번호: {password}")
print("---------------------------")
print("이 비밀번호를 복사해서 사용하세요!")
```
