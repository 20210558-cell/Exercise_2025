```python
print("=== 끝말잇기 게임 ===")

prev_word = "사과"
print(f"시작 단어: {prev_word}")

turn = 1

while True:
    curr_word = input(f"\n[{turn}턴] '{prev_word}'(으)로 시작하는 단어 입력: ")

    if curr_word == "포기":
        print("게임을 종료합니다.")
        break

    if len(curr_word) < 2:
        print("두 글자 이상 입력해야 합니다. (다시 입력해주세요.)")
        continue

    if prev_word[-1] == curr_word[0]:
        print("정답!")
        prev_word = curr_word
        turn += 1
    else:
        print(f"틀렸습니다. '{prev_word[-1]}'(으)로 시작해야 합니다.")
        print("게임 종료")
        break
```
