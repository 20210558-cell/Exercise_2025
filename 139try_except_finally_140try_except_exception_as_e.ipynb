```python
fileList = ['file01.jpg', 'file02.jpg', 'file03', 'file04.jpg']
try:
  for file in fileList:
    fname = file.split('.')         # 파일 이름과 파일확장자 분리
    print(f'파일 이름: {fname[0]}, 파일확장자: {fname[1]}')
except:
  print(f'ERROR!! <{file}> 처리중 오류 발생!')
else:
  print('모든 파일을 성공적으로 처리했습니다.')
finally:
  print('무조건 수행되는 코드!')
```

```python
intList = range(10)
idx = 10
try:
  print(intList[idx])
except Exception as e:
  exceptionClass = e.__class__.__name__
  exceptionDetail = e.__str__()          # print(e) = print(e.__str__())
  print(f'예외종류: {exceptionClass}, 예외내용: {exceptionDetail}')
```
