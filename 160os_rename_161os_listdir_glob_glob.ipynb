```python
from os import listdir

folder = './'                   # 현재 작업 디렉토리
ret = listdir(folder)
print(ret)
```

```python
from glob import glob
target = '*.txt'
ret = glob(target)
print(ret)
```

```python
target = '**'
ret = glob(target, recursive=True)
ret.sort()
for r in ret:
  print(r)
```
