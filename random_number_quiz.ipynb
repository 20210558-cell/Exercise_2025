```python
import random

computer_number = random.randint(1, 100)
max_attempts = 10

print(f"=== 숫자 맞추기 게임 (기회: {max_attempts}번 ===")

for i in range(1, max_attempts + 1):
    try:
        print(f"\n[ {i}번째 시도 ]")
        user_input = int(input("숫자를 입력하세요: "))

        if user_input < computer_number:
            print("더 큰 숫자입니다.")
        elif user_input > computer_number:
            print("더 작은 숫자입니다.")
        else:
            print(f"축하합니다! {i}번 만에 정답을 맞추셨습니다!")
            break

    except ValueError:
        print("오류: 숫자를 입력해주세요.")

else:
    print(f"\n[게임 오버] 기회를 모두 썼습니다. 정답은 {computer_number}였습니다.")
```
