```python
print("=== 글자수 세기 ===")
print("내용을 입력해주세요.")

text = input("입력 > ")

length_all = len(text)
length_no_space = len(text.replace(" ", ""))

word_list = text.split()
word_count = len(word_list)

print(f"\n총 글자 수 (공백 포함): {length_all}자")
print(f"총 글자 수 (공백 제외): {length_no_space}자")
print(f"단어 개수: {word_count}개")
```
