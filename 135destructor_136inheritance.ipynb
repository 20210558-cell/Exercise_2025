```python
class Wizard:
  def __init__(self, name=None):
    self.name = name
    print(f'마법사 이름이 <{name}>인 캐릭터가 생성되었습니다.')

  def __del__(self):
    print('Wizard 객체가 메모리에서 제거됩니다.')
```

```python
player1 = Wizard('StormBaby')
del player1     # player1 객체를 메모리에서 제거함
```

```python
class Character:
  def __init__(self):   # 생성자에서 캐릭터의 공통 능력치를 1로 초기화
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 1
    self.mentality = 1
    self.agility = 1
```

```python
class Wizard(Character):
  def setBasic(self, name=None, level=1, hp=100):
    self.name = name
    self.level = level
    self.hp = hp
```

```python
player1 = Wizard()
player1.setBasic('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

```python
class Wizard(Character):
  def __init__(self, name=None):
    super().__init__()
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```
