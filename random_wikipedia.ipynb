```python
import wikipedia
import textwrap

wikipedia.set_lang("ko")

def get_random_article():
    print("\n" + "="*40)
    print("랜덤 기사를 찾는 중입니다.")

    while True:
        try:
            title = wikipedia.random(1)

            print(f"\n[랜덤 기사]: {title}")
            choice = input("이 기사를 읽으시겠습니까? (y: 예 / n: 아니오 / s: 종료) > ").lower()

            if choice == 'y':
                page = wikipedia.page(title, auto_suggest=False)

                print("-" * 40)
                print(f"제목: {page.title}")
                print(f"URL: {page.url}")
                print("-" * 40)

                summary = textwrap.fill(page.summary, width=80)
                print(summary)
                print("-" * 40)

                input("\n[Enter]를 누르면 다음 기사를 찾습니다.")
                print("\n" + "="*40)

            elif choice == 'n':
                print("다른 기사를 찾습니다.")
                continue

            elif choice == 's':
                print("프로그램 종료")
                break

            else:
                print("다시 입력해주세요.")

        except wikipedia.exceptions.DisambiguationError:
            print("\n다른 기사를 찾습니다.")
            continue

        except wikipedia.exceptions.PageError:
            print("\n오류가 발생했습니다. 다른 기사를 찾습니다.")
            continue

        except Exception as e:
            print(f"\n오류 발생: {e}")
            break

if __name__ == "__main__":
    print("=== wikipedia 랜덤 기사 찾기 ===")
    get_random_article()
```
