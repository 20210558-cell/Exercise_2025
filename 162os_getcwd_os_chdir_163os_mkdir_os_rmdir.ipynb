```python
from os import getcwd

print(f'현재 작업 디렉토리: {getcwd()}')
```

```python
from os import chdir
print(f'현재 작업 디렉토리: {getcwd()}')
chdir('..')                                   # 상대경로 지정
print(f'현재 작업 디렉토리: {getcwd()}')
chdir('/content')                    # 절대경로 지정
print(f'현재 작업 디렉토리: {getcwd()}')
```

```python
from os import mkdir

folder = 'myfolder'
mkdir(folder)
```

```python
folder = '/content/myfolder2'
mkdir(folder)
```

```python
from os import rmdir

folder = 'myfolder'
folder2 = '/content/myfolder2'
rmdir(folder)
rmdir(folder2)
```
